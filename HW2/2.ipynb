```python
import numpy as np
import matplotlib.pyplot as plt

k =  8.617333262e-5
c = 299792458
h = 4.135667696e-15
def f(x):
    return 5*np.exp(-x) + x - 5

def binary_search(f, x1, x2, err):
    num_steps = 0
    while abs(x2 - x1) > err:
        x3 = (x1 + x2) / 2
        if f(x3)*f(x1) > 0: #same sign
            x1 = x3
        else:
            x2 = x3
        num_steps += 1
    return x2, num_steps

def b(x):
    return h * c / (k * x)

def T(x, lam):
    print("b(x)=", bx  :=  b(x), "lam=", lam)
    return bx / lam

if __name__ == "__main__":

    x_values = np.linspace(4.75, 5.25, 100)
    plt.figure(figsize=(7,5))   
    plt.plot(x_values,f(x_values), '--', label='Central difference error', color='gray')
    plt.xlabel('Step size h')
    plt.ylabel('Relative Error ε')
    plt.grid(True, which="both", ls="--")
    plt.legend()
    plt.show()
    root, steps = binary_search(f, 4, 8, 1e-6)
    print("Root is approximately:", root)
    print("Number of steps taken:", steps)

    print("The surface temperature of a star with peak wavelength 502 nm is approximately", T(root, 502e-9), "K")


    
```
