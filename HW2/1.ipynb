```python
import numpy as np

def f(x, c):
    return 1 - np.exp(-1 * c * x)

def g(x, c):
    return 1 - np.sqrt(x)

def relax(f, initial_guess, rate, err, c = 2):
    x = initial_guess
    y = (1 + rate) * f(x, c) - (rate * x)
    num_steps = 0
    while np.abs(y - x) > err:
        x = y
        y = (1 + rate) * f(x, c) - (rate * x)
        num_steps += 1
        print("y=", y, "num_steps=", num_steps)
    return num_steps

if __name__ == "__main__":
    print("We relaxed with rate=", 0, "Initial guess was 1. Number of steps was", relax(f, 1, 0, 10e-6))
    print("We relaxed with rate=", .1, "Initial guess was 1. Number of steps was", relax(f, 1, 0.1, 10e-6))
    print("We relaxed with rate=", .5, "Initial guess was 1. Number of steps was", relax(f, 1, 0.5, 10e-6))


    print("We relaxed with rate=", 0, "Initial guess was 2. Number of steps was", relax(f, 2, 0, 10e-6))
    print("We relaxed with rate=", .1, "Initial guess was 2. Number of steps was", relax(f, 2, 0.1, 10e-6))
    print("We relaxed with rate=", .5, "Initial guess was 2. Number of steps was", relax(f, 3, 0.5, 10e-6))


    print("We relaxed with rate=", 0, "Initial guess was 3. Number of steps was", relax(f, 3, 0, 10e-6))
    print("We relaxed with rate=", .1, "Initial guess was 3. Number of steps was", relax(f, 3, 0.1, 10e-6))
    print("We relaxed with rate=", .1, "Initial guess was 3. Number of steps was", relax(f, 1, 0.5, 10e-6))

    
    print("We relaxed with rate=", 0, "Number of steps was", relax(g, .5, 0, 10e-6))
    print("We relaxed with rate=", .1, "Number of steps was", relax(g, .5, -0.5, 10e-6))


```
