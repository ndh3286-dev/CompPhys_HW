```python
import numpy as np
import matplotlib.pyplot as plt

def fourier_transform(data):
    n = len(data)
    dt = 1  # Assuming uniform time steps of 1 unit
    freqs = np.fft.fftfreq(n, d=dt)
    fft_values = np.fft.rfft(data)
    return freqs, fft_values

with open("dow.txt") as f:
    data = np.loadtxt(f)
    plt.plot(data)
    plt.xlabel("Number of Days")
    plt.ylabel("Closing Dow Jones Index Value")
    plt.title("Dow Jones Industrial Average, 2006-2010")
    plt.show()
    freqs, amps = fourier_transform(data)
    # set first 10 percent of amps to 0
    plt.plot(freqs[:len(freqs)//2],  np.abs(amps[:len(amps) - 1]))
    plt.yscale('log')
    plt.xlabel("Dow Jones Index Frequency (1/day)")
    plt.ylabel("Amplitude")
    plt.title("Fourier Transform of Dow Jones Data")
    plt.show()
    amps[len(amps)//10:] = 0
    plt.plot(freqs[:len(freqs)//2], np.abs(amps[:len(amps) - 1]))
    plt.yscale('log')
    plt.xlabel("Dow Jones Index Frequency (1/day)")
    plt.ylabel("Amplitude")
    plt.title("Low-Pass Filter (10%) on Fourier Transform of Dow Jones Data")
    plt.show()

    y =  np.fft.irfft(amps, n=len(data))
    plt.plot(data, label="Original Data")
    plt.plot(y, label="Filtered Data")
    plt.xlabel("Number of Days")
    plt.ylabel("Closing Dow Jones Index Value")
    plt.title("DOW Jones Data After Low-Pass Filter (10%)")
    plt.show()

with open("dow.txt") as f:
    data = np.loadtxt(f)
    plt.plot(data)
    plt.xlabel("Number of Days")
    plt.ylabel("Closing Dow Jones Index Value")
    plt.title("Dow Jones Industrial Average Since 2006")
    plt.show()
    freqs, amps = fourier_transform(data)
    # set first 10 percent of amps to 0
    plt.plot(freqs[:len(freqs)//2],  np.abs(amps[:len(amps) - 1]))
    plt.yscale('log')
    plt.xlabel("Dow Jones Index Frequency (1/day")
    plt.ylabel("Amplitude")
    plt.title("Low-Pass Filter (10%) on Fourier Transform of Dow Jones Data")    
    plt.show()
    amps[int(len(amps)*0.02):] = 0
    plt.plot(freqs[:len(freqs)//2], np.abs(amps[:len(amps) - 1]))
    plt.plot(data, label="Original DOW Jones Data")

    plt.yscale('log')
    plt.xlabel("Dow Jones Index Frequency (1/day)")
    plt.ylabel("Amplitude")
    plt.title("Low-Pass Filter (2%) on Fourier Transform of Dow Jones Data")
    plt.show()

    y =  np.fft.irfft(amps, n=len(data))
    plt.plot(data, label="Original Data")
    plt.plot(y, label="Filtered Data")
    plt.xlabel("Number of Days")
    plt.ylabel("Closing Dow Jones Index Value")
    plt.title("Dow Jones Industrial Average - After Low-Pass Filter (2%)")
    plt.show()
```
